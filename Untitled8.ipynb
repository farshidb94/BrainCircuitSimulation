```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Load the dataset from the URL into a pandas DataFrame
url = 'https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv'
df = pd.read_csv(url)

# 2. Prepare the data: Filter for the year 2007
df_2007 = df[df['year'] == 2007].copy()

# 3. Create a color map for the continents
# This is a robust way to assign a color to each unique continent name
continents = df_2007['continent'].unique()
colors = plt.colormaps.get_cmap('viridis') # Get a color from a map
color_map = {continent: colors(i/len(continents)) for i, continent in enumerate(continents)}

# Map the continent names to the actual colors for plotting
df_2007['color'] = df_2007['continent'].map(color_map)

# 4. Set up the plot
fig, ax = plt.subplots(figsize=(12, 8))

# 5. Create the scatter plot, encoding population as size and continent as color
ax.scatter(x=df_2007['gdpPercap'],
           y=df_2007['lifeExp'],
           s=df_2007['pop'] / 100000, # Scale down population for dot size
           c=df_2007['color'],
           alpha=0.7) # Use transparency for overlapping dots

# 6. Style the plot
ax.set_xscale('log') # Use a logarithmic scale for the x-axis
ax.set_title('Wealth vs. Health of Nations in 2007', fontsize=16)
ax.set_xlabel('GDP per Capita (log scale)')
ax.set_ylabel('Life Expectancy (years)')
ax.grid(True, linestyle='--', alpha=0.6)

# 7. Create a custom legend for the continents
# We create "proxy artists" for the legend because we can't label the scatter points directly
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', label=continent,
                          markerfacecolor=color, markersize=10)
                  for continent, color in color_map.items()]

ax.legend(handles=legend_handles, title='Continent')

plt.show()
plt.close()
```
